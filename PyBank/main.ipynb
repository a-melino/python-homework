```python
"""
The following program is for the Module 2 Python Assignment.
The program deals with reading monthly profit/loss data from a .csv file and creating
a simple report as command line output.
Programmed by Alex Melino.
"""

import os    # Import the os module for getting file path
import csv   # Import the csv module for reading .csv files

current_dir = os.getcwd()   # Get current working directory

path_budget_data = os.path.join(current_dir, "budget_data.csv")   # Get the .csv file path


"""
The following function reads the data from the .csv file and returns the data as nested dictionary entries in a list. Each 
dictionary represents a row of data from the .csv file.
"""
def read_csv_file(file_path):
    data = []
    with open(file_path, 'r') as file:
        reader = csv.DictReader(file)
        for row in reader:
            data.append(dict(row))
    return data


# Calls the previous function and assigns it to a list called 'data'
data = read_csv_file(path_budget_data)


# Uses len function to determine the number of entries in the list, therefore providing the number of months in the data
month_count = len(data)


"""
The following 'for' loop does two things: it provides the sum ('profit_sum') of all dictionary values within the list 'data' 
and also creates a new list (profit_loss_list) with only the Profit/Losses values from the nested dictionary entries within 
the list 'data'
"""
profit_sum = 0
profit_loss_list = []
for entry in data:
    current_profit = int(entry['Profit/Losses'])
    profit_sum += current_profit
    profit_loss_list.append(entry['Profit/Losses'])

    
"""
The following function is used to calculate the changes in profit and loss on a month-to-month basis. It takes in a list and
returns the delta of each monthly change in a new list
"""
def get_changes(num):
    changes = []
    for i in range(1,len(num)):
        current_number = int(num[i])
        previous_number = int(num[i-1])
        change = current_number - previous_number
        changes.append(change)
    return changes


# Calls the previous function with profit_loss_list as input and assigns the output to a list called 'changes'
changes = get_changes(profit_loss_list)


# Calculates the average of the changes by dividing the sum of the list by the list length, and rounds to two digits
average_change = round(sum(changes)/len(changes),2)


# The following lines use min/max functions to determine the highest increase and decrease in changes
max_change = max(changes)
min_change = min(changes)


# The following lines find the index of the max and min changes within the initial data set 'data' and determine the 'Date' for each
max_month = data[changes.index(max(changes))+1]['Date']
min_month = data[changes.index(min(changes))+1]['Date']


# Print out all findings
print(f"\nFinancial Analysis \n{'-'*28}")
print(f"Total Months: {month_count}")
print(f"Total: ${profit_sum}")
print(f"Average Change: ${average_change}")
print(f"Greatest Increase in Profits: {max_month} (${max_change})")
print(f"Greatest Decrease in Profits: {min_month} (${min_change})")
```
